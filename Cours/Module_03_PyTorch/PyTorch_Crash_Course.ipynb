Ce notebook vous montre les 3 piliers de PyTorch :
1. Les Tenseurs (`torch.Tensor`)
2. Le calcul du gradient automatique (`Autograd`)
3. La boucle d'entraînement personnalisée (*Custom Training Loop*)

```python
import torch
import numpy as np

print("Version de PyTorch :", torch.__version__)
```

## 1. Les Tenseurs PyTorch
Un tenseur PyTorch est presque identique à un tableau NumPy, à une différence majeure : **il peut être calculé sur carte graphique (GPU)**.

```python
# Création d'un tenseur 2D (ex: batch de 2 images RGB 3x3)
x = torch.rand(2, 3, 3)
print("Tenseur :\n", x)
print("Shape :", x.shape)
print("Device :", x.device)  # Par défaut sur 'cpu'
```

## 2. Autograd (La magie du Deep Learning)

PyTorch garde en mémoire l'historique des opérations mathématiques pour calculer la dérivée (le gradient) automatiquement avec la fonction `.backward()`.

```python
# Créons un tenseur x dont on veut suivre le gradient (requires_grad=True)
x = torch.tensor(2.0, requires_grad=True)

# Faisons une opération mathématique simple : y = x^2 + 5x
y = x**2 + 5*x
print("Résultat y =", y)

# Calculons le gradient (dy/dx)
# La dérivée analytique est 2x + 5. Pour x=2, dx = 2(2)+5 = 9
y.backward()

print("Gradient dy/dx calculé par PyTorch :", x.grad)
```

## 3. La Boucle d'Entraînement (Custom Training Loop)

Contrairement à `model.fit()` dans Keras, en PyTorch on écrit explicitement les 5 étapes de l'entraînement pour chaque *batch* :

1. **Forward pass** : Prédire avec le modèle `y_pred = model(X)`
2. **Loss** : Calculer l'erreur `loss = criterion(y_pred, y_true)`
3. **Zero Grad** : Vider les gradients précédents `optimizer.zero_grad()`
4. **Backward pass** : Calculer les nouveaux gradients `loss.backward()`
5. **Optimizer Step** : Mettre à jour les poids `optimizer.step()`

```python
import torch.nn as nn
import torch.optim as optim

# Exemple minimal (Régression linéaire très basique)
model = nn.Linear(in_features=1, out_features=1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Données factices
X_train = torch.tensor([[1.0], [2.0], [3.0]])
y_train = torch.tensor([[2.0], [4.0], [6.0]]) # La relation est y = 2x

# Boucle d'entraînement pour 50 epochs
for epoch in range(50):
    # 1. Forward
    predictions = model(X_train)
    
    # 2. Loss
    loss = criterion(predictions, y_train)
    
    # 3. Zéro Grad
    optimizer.zero_grad()
    
    # 4. Backward
    loss.backward()
    
    # 5. Step (Mise à jour)
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/50], Loss: {loss.item():.4f}")

# Testons la prédiction après entraînement
test_value = torch.tensor([[4.0]])
print("\nPrédiction pour X=4 (On s'attend à ~8) :", model(test_value).item())
```
