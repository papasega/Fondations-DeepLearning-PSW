## 1. Variables et Types de Données

En Python, il n'est pas nécessaire de déclarer explicitement le type d'une variable. Le langage comprend automatiquement si c'est un entier (`int`), un nombre à virgule flottante (`float`), une chaîne de caractères (`str`) ou un booléen (`bool`).

## 2. Structures de Données : Listes et Dictionnaires

Les **listes** de Python permettent de stocker des séquences ordonnées. Les **dictionnaires** stockent les données sous forme de paires `clé: valeur` et sont très utilisés pour stocker la configuration de vos futurs réseaux de neurones.

```python
# Manipulation de Liste
ma_liste = [10, 20, 30, 40]
ma_liste.append(50) # Ajoute un élément à la fin
print("Liste modifiée :", ma_liste)
print("Le premier élément est :", ma_liste[0])

# Manipulation de Dictionnaire
parametres = {
    'epochs': 100,
    'batch_size': 32,
    'learning_rate': 0.001
}
print("Le taux d'apprentissage (learning_rate) est :", parametres['learning_rate'])
```

## 3. Boucles et Conditions

La logique conditionnelle permet d'exécuter différents blocs de code en fonction de conditions. Les boucles `for` sont essentielles pour parcourir vos jeux de données (epochs).

```python
# Boucle for (qui simule les 'epochs' d'entraînement)
for epoch in range(1, 4):  # Va de 1 à 3
    print(f"--- Epoch {epoch}/3 ---")
    
    # Condition / logique factice
    if epoch == 3:
        print("C'est la dernière époque ! L'entraînement est terminé.")
    else:
        print("Entraînement en cours...")
```

## 4. Les Fonctions

Les fonctions encapsulent le code permettant de le réutiliser. Le mot-clé est `def`.

```python
def calculer_erreur_absolue(y_vrai, y_pred):
    """Calcule l'erreur absolue entre la vraie valeur et la valeur prédite."""
    erreur = abs(y_vrai - y_pred)
    return erreur

vrai = 5.0
prediction = 4.2
print(f"L'erreur absolue est de : {calculer_erreur_absolue(vrai, prediction)}")
```

## 5. NumPy : Le Moteur Mathématique

En Machine / Deep Learning, une liste Python standard est trop lente pour traiter des millions de données. C'est là que **NumPy** intervient, avec ses "tableaux à N dimensions" (*Ndarrays* ou *Tensors*).

```python
import numpy as np

# Création d'une matrice (tableau 2D)
matrice = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Matrice :\n", matrice)
print(f"Dimensions (Shape) : {matrice.shape} -> 2 lignes, 3 colonnes")

# Opérations mathématiques optimisées
matrice_carre = matrice ** 2
print("\nMatrice élevée au carré :\n", matrice_carre)
print("Moyenne de la matrice :", np.mean(matrice))
```

##  6. Matplotlib : Visualisation des données

La visualisation est cruciale pour comprendre le fonctionnement de nos modèles ou la géométrie de nos données.

```python
import matplotlib.pyplot as plt

# Génération de points entre -10 et 10 via NumPy
x = np.linspace(-10, 10, 100)
y = x ** 2  # Une simple parabole (ex: fonction cost en régression linéaire)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label='Courbe f(x) = x²')
plt.title('Traçons une fonction mathématique')
plt.xlabel('Axe des X')
plt.ylabel('Axe des Y')
plt.axhline(0, color='black', linewidth=0.5) # Ligne x=0
plt.axvline(0, color='black', linewidth=0.5) # Ligne y=0
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
