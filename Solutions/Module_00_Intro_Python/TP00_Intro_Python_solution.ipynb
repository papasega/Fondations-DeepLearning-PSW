```python
import matplotlib.pyplot as plt

epochs = np.arange(1, 11)
loss = np.exp(-0.3 * epochs) + np.random.normal(0, 0.05, 10)

plt.figure(figsize=(8, 5))
plt.plot(epochs, loss, marker='o', linestyle='-', color='red', label='Training Loss')
plt.title('Évolution de la Perte (Loss) pendant l\'entraînement')
plt.xlabel('Époques (Epochs)')
plt.ylabel('Perte (Loss)')
plt.grid(True)
plt.legend()
plt.show()
```
