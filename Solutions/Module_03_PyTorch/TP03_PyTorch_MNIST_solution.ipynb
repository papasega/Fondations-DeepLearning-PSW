# TP : Classification MNIST avec PyTorch (Solution)

Solution complète du TP de classification MNIST avec PyTorch.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# 1. Chargement des données MNIST avec torchvision
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)) # Normalisation basique
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# SOLUTION 1 : Création des DataLoaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

## 2. Définition du Modèle (DNN) PyTorch

```python
class DNN_MNIST(nn.Module):
    def __init__(self):
        super(DNN_MNIST, self).__init__()
        # SOLUTION 2 : Définition des couches
        self.fc1 = nn.Linear(28 * 28, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        # SOLUTION 3 : Forward pass
        x = x.view(-1, 28 * 28) # Flatten (batch_size, 784)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        # Pas de Softmax car utiliser par CrossEntropyLoss
        return x
```

## 3. Initialisation et Boucle d'entrainement

```python
model = DNN_MNIST()

# SOLUTION 4 : Initialisation
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 3
print("Début de l'entraînement...")

for epoch in range(epochs):
    running_loss = 0.0
    for images, labels in train_loader:
        # SOLUTION 5 : Les 5 étapes de la boucle
        # 1. Forward
        outputs = model(images)
        # 2. Loss
        loss = criterion(outputs, labels)
        # 3. Zéro Grad
        optimizer.zero_grad()
        # 4. Backward
        loss.backward()
        # 5. Step
        optimizer.step()
        
        running_loss += loss.item()
        
    print(f"Epoch {epoch+1}/{epochs} - Loss: {running_loss/len(train_loader):.4f}")
```
